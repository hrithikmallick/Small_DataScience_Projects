```python
#Load all required librayies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sys
```

```python
marvelData=pd.read_csv("./charcters_stats.csv")
marvelData.head()
```

```python
marvelData.shape
```

```python
#Sort the data in case of speed
marvelData.sort_values(by="Speed",ascending=False).head()
```

```python
#Sort the data in case of speed
marvelData.sort_values(by="Strength",ascending=False).head()
```

```python
badChar=marvelData[marvelData["Alignment"]=="bad"][0:10]
```

```python
badChar.head(10)
```

```python
#sorted bad char
bad_max_pwr=badChar.sort_values(by="Total",ascending=False)
bad_max_pwr.head()
```

```python
#valus count of allignment
marvelData["Alignment"].value_counts()
```

```python
#Good character
goodChar=marvelData[marvelData["Alignment"]=="good"]
goodChar.sort_values(by="Power",ascending=False).head()
```

```python
#find all suprer heros whose power is above 90
good_pwr=goodChar[goodChar["Power"]>=90]
good_pwr.head()
```

```python
good_pwr.shape
```

```python
good_pwr
```

```python
good_max_pwr=goodChar.sort_values(by="Total",ascending=False)
good_max_pwr.head()
```

```python
#make bar plot of suprer heros with max power
plt.figure(figsize=(17,5))
plt.grid()
plt.bar(list(good_max_pwr["Name"])[0:10],list(good_max_pwr["Total"])[0:10])
plt.show()
```

```python
#make bar plot of suprer vilans with max power
plt.figure(figsize=(17,5))
plt.grid()
plt.bar(list(bad_max_pwr["Name"])[0:5],list(bad_max_pwr["Total"])[0:5])
plt.show()
```

```python

#Makingahistogram for speed for good superheroes
plt.figure(figsize=(10,5))
plt.hist(goodChar['Speed'])
plt.title("Distribution of speed")
plt.xlabel("Speed")
plt.show()
```
