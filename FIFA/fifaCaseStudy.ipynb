```python
#Loading all required library's
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sys
```

```python
fifaData=pd.read_csv("./players_20.csv")
fifaData.head()
```

```python
fifaData.shape
```

```python
#Finding by short name
fifaData[fifaData["short_name"]=="L. Messi"]
```

```python
#Print every column name
for i in fifaData.columns:
    print(i)
```

```python
#Experity data analysis
fifaData["nationality"].value_counts()[0:10]
```

```python
fifaData["club_name"].value_counts()[0:10]
```

```python
#make bar plot of top 10 nationality 
plt.figure(figsize=(17,7))
plt.grid()
plt.bar(list(fifaData["nationality"].value_counts()[0:10].keys()),list(fifaData["nationality"].value_counts()[0:10]))
plt.show()
```

```python
#player salary and short name 
player_salary=fifaData[["short_name","wage_eur"]]
player_salary.head()
```

```python
#sort a column from a dataset
sort_player=player_salary.sort_values(by=["wage_eur"],ascending=False)
sort_player.head(10)
```

```python
#Bar plot for top 5 salary plaeyers
plt.figure(figsize=(7,5))
plt.bar(sort_player["short_name"][0:5],sort_player["wage_eur"][0:5],color=["brown","blue","yellow","red","purple"])
plt.grid()
plt.show()
```
